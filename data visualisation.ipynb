* to draw the distribution plot to check how the data is distrubuted

```python
sns.distplot(data["normalized_losses"])
plt.show()
```

```python
sns.distplot(data["city_mpg"], kde= False, rug= True)
plt.show()
```

```python
sns.distplot(data["normalized_losses"], kde= False, rug= True)
plt.show()
```

```python
# this is the joint plot between two variables
sns.jointplot(data["engine_size"], data["horsepower"])
plt.show()
```

```python
# put hue to check where the data is highly densed
sns.jointplot(data["engine_size"], data["horsepower"], kind = "hex")
plt.show()
```

```python
# residual plot between two variables 
sns.jointplot(data["engine_size"], data["horsepower"], kind = "resid")
plt.show()
```

```python
#density plot
sns.jointplot(data["engine_size"], data["horsepower"], kind = "kde")
plt.show()
```

```python
data.head()
```

```python
#strip plot to visualize categorical variables

sns.stripplot(data["fuel_type"], data["horsepower"])
plt.show()
```

```python
# swarm plot is much better then strip plot.
sns.swarmplot(data["fuel_type"], data["horsepower"])
plt.show()
```

```python
# boxplot for categorical variables
sns.boxplot(data["number_of_doors"], data["horsepower"])
```

```python
# we can put hue to add another categorcal variable to visualize  more complex data the data
sns.boxplot(data["number_of_doors"], data["horsepower"], hue= data["fuel_type"])
```

```python
# count plot used to check the number of total counts of cat. variables in data
sns.countplot(data["body_style"])
```

```python
sns.countplot(data["body_style"], hue= data["engine_location"])
```

```python
# we can use factor plot to plot different types of plots
sns.factorplot(x="number_of_doors", y="engine_size", data= data, kind= "violin")
```

```python
sns.factorplot(x="number_of_doors", y="engine_size", data= data, kind= "box")
```

```python
sns.factorplot(x="number_of_doors", y="engine_size", data= data, kind= "strip")
```

```python
sns.factorplot(x="number_of_doors", y="engine_size", data= data, kind= "swarm")
```

```python
# scatter plot shows the relationship between two variables
plt.scatter(x= "stroke", y="horsepower", data= data)
```
